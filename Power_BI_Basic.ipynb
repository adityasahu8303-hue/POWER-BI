Question 1 : What is Power BI and why is it used in businesses?
-- Power BI is a data visualization and business intelligence (BI) tool that converts raw data into meaningful insights using charts, graphs, and dashboards.
*Businesses use Power BI for several important reasons:*

1. Better Decision Making
2. Data Visualization
3. Real-Time Insights
4. Data Integration
5. Improved Efficiency
6. Collaboration & Sharing

Question 2 : Name and explain the three main components of Power BI.
--
1. Power BI Desktop
It is a Windows application used to create reports and dashboards.
You can connect, clean, and transform data from different sources.
It allows you to design visualizations like charts, graphs, and tables.
2. Power BI Service
It is a cloud-based platform (online).
Used to publish, share, and access reports from anywhere.
Supports collaboration among team members.
3. Power BI Mobile
A mobile application available on phones and tablets.
Lets users view dashboards and reports on the go.
Provides real-time updates and notifications.

Question 3 : Explain the Power BI workflow.
1. Data Collection (Extract)
Data is collected from different sources like Excel, databases, APIs, etc.
Power BI can connect to many sources directly.

2. Data Transformation (Transform)
Data is cleaned and prepared using Power Query.
Tasks include removing duplicates, handling missing values, formatting columns, etc.

3. Data Modeling
Relationships are created between tables.
Calculations are added using DAX (Data Analysis Expressions).

4. Data Visualization (Report Creation)
Create charts, graphs, and dashboards in Power BI Desktop.
Use visuals like bar charts, pie charts, tables, etc.

5. Publishing
Reports are published to Power BI Service.
Makes reports available online.

6. Sharing & Collaboration
Reports and dashboards are shared with team members.
Users can collaborate and view insights.

7. Monitoring & Refresh
Data is refreshed regularly (automatic/manual).
Users monitor dashboards for updates.

Question 5 : Write step by step instructions to load the above dataset.
1. Open Power BI Desktop
    . Launch Power BI Desktop on your computer.
2. Click on “Get Data”
    . On the Home tab, click Get Data.
    . Choose the data source (e.g., Excel, CSV, SQL Server).
3. Select the File
    . Browse your computer and select the dataset file.
Click Open.
4. Preview the Data
    . A preview window will appear showing the dataset.
    . Check if the data looks correct.
5. Click “Load” or “Transform Data”
    . Click Load to directly import data.
    . OR click Transform Data to clean/modify data using Power Query.
6. Data Loaded Successfully
    . The dataset will appear in the Fields pane.
    . Now you can start creating reports and visuals.

Question 6 : Define Data View, Report View, and Model View. Explain the purpose of each view.
--
1. Data View
Displays the actual data in table format (rows and columns).
Allows you to inspect, filter, and create calculated columns.

    1.Purpose: To view and understand raw data.

2. Report View
Used to create visualizations like charts, graphs, maps, and dashboards.
You can design interactive reports using drag-and-drop features.

    1.Purpose: To analyze data and present insights visually.

🔗 3. Model View
Shows relationships between different tables in a diagram format.
Allows you to create and manage relationships (one-to-many, etc.).

    1.Purpose: To organize and manage data structure and relationships

Question 7 : Discuss the different data sources that Power BI supports.
--
1. File-Based Sources
Data stored in files on your system.
Examples:
    1.Excel (.xlsx)
    2.CSV (.csv)
    3.Text files

 Use: When data is stored locally.

2. Database Sources
Connects to structured databases.
Examples:
    1.SQL Server
    2.MySQL
    3.Oracle
    4.PostgreSQL

 Use: For large and structured business data.

3. Cloud-Based Sources
Data from online/cloud services.
Examples:
    1.Azure SQL Database
    2.Google Analytics
    3.SharePoint
    4.OneDrive

 Use: For real-time and cloud-stored data.

4. Web & APIs
Fetch data from websites or APIs.
Example:
    1.Web pages (URL)
    2.REST APIs
  Use: For live or external data integration.

Question 8 : Split Owner Name to create two new columns as First Name and Last Name
--
Steps to Split Owner Name
1. Open Power Query
In Power BI Desktop, click Transform Data.
2. Select the “Owner Name” Column
Click on the column that contains full names (e.g., Ram Sharma).
3. Use Split Column Option
Go to Home or Transform tab
Click Split Column → By Delimiter
4. Choose Delimiter
Select Space (" ") as the delimiter
Choose Split at Left-most delimiter
5. Rename Columns
Rename:
First column → First Name
Second column → Last Name
6. Apply Changes
Click Close & Apply to save changes.
